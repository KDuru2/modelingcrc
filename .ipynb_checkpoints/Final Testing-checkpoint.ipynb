## Data Initialization

```python
import pandas as pd
import numpy as np
import torch as T
from sklearn.metrics import roc_curve, auc, roc_auc_score
from sklearn.metrics import precision_recall_curve, f1_score, accuracy_score, precision_score, recall_score
from sklearn.svm import LinearSVC
import matplotlib.pyplot as plt
import shap
from sklearn.metrics import accuracy_score
import xgboost as xgb
```

## Model Analysis

```python
#Loads saved data
trainXdf = pd.read_csv('nfinaltrainX.csv')
testXdf = pd.read_csv('nfinaltestX.csv')
validXdf = pd.read_csv('nfinalvalidX.csv')
trainYdf = pd.read_csv('ntrainY.csv')
testYdf = pd.read_csv('ntestY.csv')
validYdf = pd.read_csv('nvalidY.csv')

#Creates tensors
trainX = trainXdf.to_numpy()
testX = testXdf.to_numpy()

trainY = trainYdf.to_numpy()
testY = testYdf.to_numpy()

validX = validXdf.to_numpy()
validY = validYdf.to_numpy()

TrainX = T.tensor(trainX, dtype=T.float32)
TrainY = T.tensor(trainY, dtype=T.float32).reshape(-1, 1)

TestX = T.tensor(testX, dtype=T.float32)
TestY = T.tensor(testY, dtype=T.float32).reshape(-1, 1)

ValidX = T.tensor(validX, dtype=T.float32)
ValidY = T.tensor(validY, dtype=T.float32).reshape(-1, 1)
```

```python
from sklearn.metrics import roc_curve, auc
analysis_model = T.load('nfinal_model.pth', weights_only=False)

analysis_model.eval()

with T.no_grad():
    outputs = analysis_model(TestX)
    y_probs = T.sigmoid(outputs).cpu().numpy().flatten()

    
    # Calculate FPR, TPR, and thresholds
    fpr, tpr, thresholds = roc_curve(TestY, y_probs)
    
    # Calculate the AUC (Area Under the Curve)
    aucreport = auc(fpr, tpr)

    # Plot the ROC curve
    plt.figure()
    plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.4f)' % aucreport)
    plt.fill_between(fpr, tpr, color='blue', alpha=0.05)
    plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('PyTorch Model ROC Curve')
    plt.legend(loc="lower right")
    plt.show()
```

```python
# Precision, Recall
with T.no_grad():
    outputs = analysis_model(TestX)
    y_probs = T.sigmoid(outputs).cpu().numpy().flatten()
    y_preds = (y_probs > 0.5).astype(int)  # Convert probabilities to binary predictions


# Calculate precision, recall, and thresholds
precision, recall, thresholds = precision_recall_curve(testY, y_probs)

# Calculate the AUC (Area Under the Curve) for the Precision-Recall curve
pr_auc = auc(recall, precision)

# Plot the Precision-Recall curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, color='blue', label=f'Precision-Recall AUC = {pr_auc:.4f}')
plt.fill_between(recall, precision, color='blue', alpha=0.05)
plt.ylim(0, 1.05)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc="lower left")
plt.grid(True)
plt.show()


# Calculate F1 score
f1 = f1_score(testY, y_preds)

accuracy = accuracy_score(testY, y_preds)

precision = precision_score(testY, y_preds)
recall = recall_score(testY, y_preds)

# Print the results
print(f'F1 Score: {f1:.4f}')
print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Assuming your model is already in evaluation mode
analysis_model.eval()

# Using torch.no_grad() to avoid tracking gradients
with T.no_grad():
    # Get raw model output (logits)
    logits = analysis_model(TestX)

    # Apply sigmoid to get probabilities for the positive class
    y_prob = T.sigmoid(logits).cpu().numpy().flatten()  # Convert to NumPy array

    # Convert probabilities to binary predictions (0 or 1) using threshold of 0.5
    y_preds = (y_prob > 0.5).astype(int)  # Predicted class (0 or 1)

# Compute confusion matrix
cm = confusion_matrix(TestY, y_preds)

# Display confusion matrix using ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Negative", "Positive"])
disp.plot(cmap=plt.cm.Blues)

# Show the plot
plt.title('Confusion Matrix')
plt.show()

# Optionally, print the confusion matrix
print("Confusion Matrix:")
print(cm)
```

## Comparative Models

```python
from sklearn.svm import SVC

svc_model = SVC(kernel='rbf', C=1, gamma='auto', random_state=24, probability = True)  # RBF kernel (default)
# Train model
svc_model.fit(trainX, trainY)

y_pred = svc_model.predict(testX)  # Class predictions (0 or 1)
y_probs = svc_model.predict_proba(testX)[:, 1]  # Probabilities for class 1


accuracy = accuracy_score(testY, y_pred)
precision = precision_score(testY, y_pred)
recall = recall_score(testY, y_pred)
f1 = f1_score(testY, y_pred)

print("SVC")
print(f"Accuracy: {accuracy:.3f}")
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1-score: {f1:.3f}")

aucreport1 = roc_auc_score(testY, y_probs)  # Use predicted probabilities


# Compute ROC curve
fpr1, tpr1, _ = roc_curve(testY, y_probs)


from sklearn.neural_network import MLPClassifier

mlp_model = MLPClassifier(hidden_layer_sizes=(100,), activation='relu', solver='adam', 
                          alpha=0.0, max_iter=500, random_state=42)

# Train model
mlp_model.fit(trainX, trainY)

y_pred = mlp_model.predict(testX)  # Class predictions (0 or 1)
y_probs = mlp_model.predict_proba(testX)[:, 1]  # Probabilities for class 1

accuracy = accuracy_score(testY, y_pred)
precision = precision_score(testY, y_pred)
recall = recall_score(testY, y_pred)
f1 = f1_score(testY, y_pred)

print("MLP")
print(f"Accuracy: {accuracy:.3f}")
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1-score: {f1:.3f}")

aucreport2 = roc_auc_score(testY, y_probs)  # Use predicted probabilities

#print(f"AUC Score: {aucreport:.3f}")

# Compute ROC curve
fpr2, tpr2, _ = roc_curve(testY, y_probs)


from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

adaboost_model = AdaBoostClassifier(
    n_estimators=100,
    learning_rate=1.3,
    random_state=42
)

# Train model
adaboost_model.fit(trainX, trainY)

y_pred = adaboost_model.predict(testX)  # Class predictions (0 or 1)
y_probs = adaboost_model.predict_proba(testX)[:, 1]  # Probabilities for class 1

accuracy = accuracy_score(testY, y_pred)
precision = precision_score(testY, y_pred)
recall = recall_score(testY, y_pred)
f1 = f1_score(testY, y_pred)

print("AdaBoost")
print(f"Accuracy: {accuracy:.3f}")
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1-score: {f1:.3f}")

aucreport3 = roc_auc_score(testY, y_probs)  # Use predicted probabilities

#print(f"AUC Score: {aucreport3:.3f}")

# Compute ROC curve
fpr3, tpr3, _ = roc_curve(testY, y_probs)


from sklearn.neighbors import KNeighborsClassifier
# Create K-Neighbors Classifier
model = KNeighborsClassifier(n_neighbors=2, metric='minkowski', p=25)  # 5 neighbors, Euclidean distance

# Train model
model.fit(trainX, trainY)

# Predictions
y_pred = model.predict(testX)  # Class predictions (0 or 1)
y_probs = model.predict_proba(testX)[:, 1]  # Probabilities for class 1

accuracy = accuracy_score(testY, y_pred)
precision = precision_score(testY, y_pred)
recall = recall_score(testY, y_pred)
f1 = f1_score(testY, y_pred)

print("K-Neigh")
print(f"Accuracy: {accuracy:.3f}")
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1-score: {f1:.3f}")

aucreport4 = roc_auc_score(testY, y_probs)  # Use predicted probabilities

#print(f"AUC Score: {aucreport4:.3f}")

# Compute ROC curve
fpr4, tpr4, _ = roc_curve(testY, y_probs)


from sklearn.naive_bayes import BernoulliNB

model = BernoulliNB()
# Train model
model.fit(trainX, trainY)

y_pred = model.predict(testX)  # Class predictions (0 or 1)
y_probs = model.predict_proba(testX)[:, 1]  # Probabilities for class 1

accuracy = accuracy_score(testY, y_pred)
precision = precision_score(testY, y_pred)
recall = recall_score(testY, y_pred)
f1 = f1_score(testY, y_pred)

print("BNB")
print(f"Accuracy: {accuracy:.3f}")
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1-score: {f1:.3f}")

aucreport5 = roc_auc_score(testY, y_probs)  # Use predicted probabilities

#print(f"AUC Score: {aucreport5:.3f}")

# Compute ROC curve
fpr5, tpr5, _ = roc_curve(testY, y_probs)


from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

# Train logistic regression model
model = LogisticRegression(None)
model.fit(trainX, trainY)

y_pred = model.predict(testX)  # Class predictions (0 or 1)
y_probs = model.predict_proba(testX)[:, 1]  # Probabilities for class 1

accuracy = accuracy_score(testY, y_pred)
precision = precision_score(testY, y_pred)
recall = recall_score(testY, y_pred)
f1 = f1_score(testY, y_pred)

print("Logistic R")
print(f"Accuracy: {accuracy:.3f}")
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1-score: {f1:.3f}")

aucreport6 = roc_auc_score(testY, y_probs)  # Use predicted probabilities

#print(f"AUC Score: {aucreport6:.3f}")

# Compute ROC curve
fpr6, tpr6, _ = roc_curve(testY, y_probs)


import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score


# Define XGBoost classifier
model = xgb.XGBClassifier(objective='binary:logistic', learning_rate = 0.25, n_estimators = 100)

# Train the model
model.fit(trainX, trainY)

y_pred = model.predict(testX)  # Class predictions (0 or 1)
y_probs = model.predict_proba(testX)[:, 1]  # Probabilities for class 1

accuracy = accuracy_score(testY, y_pred)
precision = precision_score(testY, y_pred)
recall = recall_score(testY, y_pred)
f1 = f1_score(testY, y_pred)

print("XGBoost")
print(f"Accuracy: {accuracy:.3f}")
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1-score: {f1:.3f}")

aucreport7 = roc_auc_score(testY, y_probs)  # Use predicted probabilities

#print(f"AUC Score: {aucreport7:.3f}")

# Compute ROC curve
fpr7, tpr7, _ = roc_curve(testY, y_probs)

# Plot ROC curve
plt.figure(figsize=(9, 6))
plt.plot(fpr, tpr, color='pink', label=f'PyTorch (AUC = {aucreport:.3f})')
plt.plot(fpr2, tpr2, color='orange', label=f'MLP (AUC = {aucreport2:.3f})')
plt.plot(fpr3, tpr3, color='green', label=f'AdaBoost (AUC = {aucreport3:.3f})')
plt.plot(fpr6, tpr6, color='cyan', label=f'Logistic Regression (AUC = {aucreport6:.3f})')
plt.plot(fpr5, tpr5, color='red', label=f'Bernoulli Naive Bayes (AUC = {aucreport5:.3f})')
plt.plot(fpr1, tpr1, color='blue', label=f'SVC (AUC = {aucreport1:.3f})')
plt.plot(fpr7, tpr7, color='brown', label=f'Stochastic Gradient Boosting (AUC = {aucreport7:.3f})')
plt.plot(fpr4, tpr4, color='purple', label=f'KNeighbors (AUC = {aucreport4:.3f})')



plt.plot([0, 1], [0, 1], linestyle='--', color='gray')  # Diagonal line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Comparative ROC Analysis")
plt.legend()
plt.show()
```
